```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Canada_income.csv")
df
```

```python
plt.xlabel("year")
plt.ylabel("income")
plt.scatter(df.year,df.income)
```

```python
new_df = df.drop("income",axis="columns")
new_df
```

```python
model = linear_model.LinearRegression()
model.fit(new_df,df.income)
```

```python
model.predict([[2020]])
```

```python
# y = m*x+b -> formula
# y -> dependent Variable  
# m -> coefficient
# x -> Independent variable
# b -> Intercept
```

```python
model.coef_
```

```python
model.intercept_
```
